```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Adjusting the colors to be more fitting for a scientific context
df = pd.read_excel('Data/External Figure Data/AI Funding Data.xlsx')

# Converting billions to actual values for plotting
df['AI-Related Funding'] = df['AI-Related Funding'] /1e7
df['Non-AI Funding'] = df['Non-AI Funding'] / 1e7

# Setting the font to Times New Roman
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'

# Creating a figure and a set of subplots
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plotting a stacked bar chart for the funding data
ax1.bar(df['Year'], df['AI-Related Funding'], width=0.4, label='AI-Related Funding', color='darkgreen', edgecolor = 'black')
ax1.bar(df['Year'], df['Non-AI Funding'], width=0.4, bottom=df['AI-Related Funding'], label='Non-AI Funding', color='darkgrey', edgecolor = 'black')

# Setting the labels for the axes
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Funding Amount (Billions USD)', fontsize=14)

# Formatting the y-axis to show labels in billions
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x/1e3:.0f}B'))

# Creating a secondary y-axis to plot the number of rounds
ax2 = ax1.twinx()
ax2.plot(df['Year'], df['Total Non-AI Rounds'], marker='o', label='Total Non-AI Rounds', color='grey', markeredgecolor='black', linewidth=2)
ax2.plot(df['Year'], df['Total AI Rounds'], marker='o', label='Total AI Rounds', color='green', markeredgecolor='black', linewidth=2)
ax2.set_ylabel('Number of Rounds', fontsize=14)

# Adding a title and a grid to the plot
plt.title('Funding To AI-Related US Startups', fontsize=16)
ax1.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5)
ax2.grid(False)

# Adding legends for both y-axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Setting the figure background to white
fig.patch.set_facecolor('white')

# Showing the plot
fig.savefig('Plots/External Figures/Funding to AI-Related US Startups.png' , dpi=300, bbox_inches='tight')
plt.show()

```
