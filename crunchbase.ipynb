## Checking the full dataframe for NAs
### General NA understanding
The goal is to get a better understanding of the dataframe structure and possible correlations that influence the sampling strategy.

```python
# List of columns to check for NAs
columns_to_check = [
    'Number of Funding Rounds',
    'Funding Status',
    'Last Funding Date',
    'Last Funding Amount Currency (in USD)',
    'Last Funding Type',
    'Total Funding Amount Currency (in USD)'
]

# Counting NAs per row in the specified columns
cb_2['NA Count'] = cb_2[columns_to_check].isna().sum(axis=1)

# Converting CB Rank (Company) to numeric
cb_2['CB Rank (Company)'] = pd.to_numeric(cb_2['CB Rank (Company)'].str.replace(',', ''), errors='coerce')

# Calculating and printing the correlation
correlation = cb_2['CB Rank (Company)'].corr(cb_2['NA Count'])
print(f"The correlation between 'CB Rank (Company)' and the amount of NAs is: {correlation}")
```

```python
def calculate_and_print_missing_values_info(df):
    """
    Calculates and prints the average number of missing values per row and
    the average percentage of missing values per row for a given DataFrame.

    Parameters:
    - df: pd.DataFrame
        The DataFrame for which missing values information is calculated.
    """
    # Calculating total missing values per row
    total_missing_values = df.isna().sum(axis=1)
    
    # Calculating average missing values per row
    average_missing_per_row = total_missing_values.mean()
    
    # Calculating missing values per row as percentages
    missing_values_percent = (total_missing_values / len(df.columns)) * 100
    
    # Calculating the average of the missing values percentages
    average_missing_percent = missing_values_percent.mean()
    
    # Printing results
    print(f"Average missing values per row: {average_missing_per_row}")
    print(f"Average missing values per row in percent: {average_missing_percent}%")
```

```python
def remove_na_rows(df):
    """
    Drops the rows with NAs in specified columns and returns a cleaned dataframe 
    Parameters:
    - df: pd.DataFrame
        The DataFrame specific rows with NAs are removed from
    """
    # Initial number of rows
    initial_row_count = len(df)
    
    # Columns to check for NAs
    columns_to_check = ["Website", "Industry Groups", 'Number of Employees'
                        ]
    
    # Dropping rows with any NAs in the specified columns
    cleaned_df = df.dropna(subset=columns_to_check)
    
    # Final number of rows
    final_row_count = len(cleaned_df)
    
    # Calculating the number of rows removed
    rows_removed = initial_row_count - final_row_count
    
    print(f"Rows removed: {rows_removed}")
    print(f"Rows remaining: {final_row_count}")
    
    return cleaned_df
```

### Removing already crawled websites from the random sample

```python
# Importing companies from the previous scrape 
scrape_1 = pd.read_csv('Data/Crunchbase/companies_work.csv')
scrape_1 = scrape_1.rename(columns = {'Website' : 'website'})
scrape_1.info()
```

```python
# Finding overlap between first scrape and new second scrape
overlap_websites = pd.merge(scrape_1[['website']], rand_sample[['website']], on = 'website', how = 'inner')['website'].unique()

# Marking Overlap in scrape_1
rand_sample['already_scraped'] = rand_sample['website'].isin(overlap_websites)
scrape_1['overlapping'] = scrape_1['website'].isin(overlap_websites)

# Removing the overlap from the second scrape, for quicker scraping, and for later reattachment
rand_sample_new = rand_sample[~rand_sample['website'].isin(overlap_websites)]
rand_sample_new.to_csv('Data/Crunchbase/rand_sample_new.csv')
rand_sample_new.info()
```

### Setting up the files to be ready to crawl
As the dataset is quite large, I divide it into batches to increase stability when scraping

```python
# Setting up the batches 
chunk_size = 1000
number_of_chunks = len(rand_sample_new) // chunk_size + (1 if len(rand_sample_new) % chunk_size else 0)

for i in range(number_of_chunks):
    # Selecting the specific chunk based on the current iteration
    start_index = i * chunk_size
    end_index = start_index + chunk_size
    websites = rand_sample_new.iloc[start_index:end_index]
    
    # Dropping rows where 'website' is NaN
    websites.dropna(subset=['website'], inplace=True)
    
    # Marking 'snapshot_in_window' as False for all rows in the current chunk
    websites['snapshot_in_window'] = False
    
    # Defining the filename with the upper end of the sample range included
    filename = f'./Data/Crunchbase full/website_{end_index}.csv'
    
    # Exporting the current chunk to a CSV file
    websites.to_csv(filename, index=False)

    print(f'Exported chunk {i+1}/{number_of_chunks} to {filename}')
```
